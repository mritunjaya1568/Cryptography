```python
from socket import socket
import string

brute_force = "{" + "}" + "_" + "@" + string.digits + string.ascii_lowercase + string.ascii_uppercase 

flag = "flag{"

while True:
    for i in brute_force:
        sock = socket()
        sock.connect(('challenges.ctfd.io', 30468))
        sock.send('{}\n'.format(flag + i).encode())

        print("Trying " + flag + i)
        if "seem to know" not in str(sock.recv(1024)):
            pass

        else:
            print("> " + i)
            flag += i
            break 
```
